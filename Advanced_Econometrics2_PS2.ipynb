```python
import pandas as pd
import numpy as np
import plotnine as p

import statsmodels.api as sm
import statsmodels.formula.api as smf
from linearmodels import IV2SLS 

from stargazer.stargazer import Stargazer
```

# part b- OLS model and 2SLS model

```python
#OLS
ols_reg = sm.OLS.from_formula("LW ~ S + IQ + EXPR + TENURE + RNS + SMSA + YEAR_66 +	YEAR_67 +	YEAR_68 +	YEAR_69 +	YEAR_70	+ YEAR_71 +	YEAR_73", 
              data = df3).fit()

print(ols_reg)
ols_reg.summary()
```

```python
#2SLS
iv_reg = IV2SLS.from_formula("LW ~ S + EXPR + TENURE + RNS + SMSA + YEAR_66 +	YEAR_67 +	YEAR_68 +	YEAR_69 +	YEAR_70	+ YEAR_71 +	YEAR_73 + [IQ ~ MED + KWW + MRT + AGE ] ", df3).fit()
print(iv_reg)
```

```python
table32 = pd.DataFrame({'A': np.linspace(1,2,2), 'Estimation technique': ['OLS','2SLS'],'S coeficient': ['0.0620','0.0692'],'IQ coeficient': ['0.0027','0.0020']})
table32
```

# part c- sargan test

```python
print(iv_reg.sargan)
```

# part d- running 2 OLS estimation

```python
res_first = IV2SLS(df3.IQ, df3[["MED" ,"KWW","MRT","AGE"]], None, None).fit(
    cov_type="unadjusted"
)
print(res_first)
df3["IQ_hat"] = df3.IQ - res_first.resids
```

```python
res_second = IV2SLS((df3.LW), df3[["S","IQ_hat","EXPR","TENURE" , "RNS",  "SMSA" , "YEAR_66" ,	"YEAR_67" ,	"YEAR_68" ,	"YEAR_69" ,	"YEAR_70"	, "YEAR_71" ,	"YEAR_73"]],None,None).fit(
    cov_type="unadjusted"
)
print(res_second)
```

# part e_ schooling is endogenous as well

```python
#2SLS
 
iv_reg2 = IV2SLS.from_formula("LW ~ EXPR + TENURE + RNS + SMSA + YEAR_66 +	YEAR_67 +	YEAR_68 +	YEAR_69 +	YEAR_70	+ YEAR_71 +	YEAR_73 + [IQ + S ~ MED + KWW + MRT + AGE ] ", df3).fit()
print(iv_reg2)
```

```python
print(iv_reg2.sargan)
```

```python
#He gives three reasons for the endogeneity of schooling. 
#First, the measurement error etha may be related to success in schooling. 
#The second reason is the argument we made in the text: 
#the choice of S is influenced by unobservable characteristics. 
#Third, schooling may be measured with error. The general formula for the asymptotic bias for the 2SLS estimator is given by:

#If S is endogenous, the instrument set xi for the 2SLS estimation in (b) erroneously
#includes a variable (which is S) that is not predetermined. So one of
#the elements of E(xi.ri) is not zero. Then it is clear from the formula that
#asymptotic bias can exist at least for some coefficients.
```

# part f - GMM

```python
from linearmodels.iv.model import IVGMM
```

```python
df3["S_hat"] = df3.S - iv_reg.resids
```

```python
#GMM without schooling as instrument
gmmest0 = IVGMM((df3.LW), df3[["EXPR","TENURE" , "RNS",  "SMSA" , "YEAR_66" ,	"YEAR_67" ,	"YEAR_68" ,	"YEAR_69" ,	"YEAR_70"	, "YEAR_71" ,	"YEAR_73"]],df[["IQ"]],df3[["MED" ,"KWW","MRT","AGE"]]).fit(
    cov_type="unadjusted"
)
print(gmmest0)
```

```python
#GMM with schooling as instrument
gmmest1 = IVGMM((df3.LW), df3[["EXPR","TENURE" , "RNS",  "SMSA" , "YEAR_66" ,	"YEAR_67" ,	"YEAR_68" ,	"YEAR_69" ,	"YEAR_70"	, "YEAR_71" ,	"YEAR_73"]],df[["IQ"]],df3[["MED" ,"KWW","MRT","AGE","S_hat"]]).fit(
    cov_type="unadjusted"
)
print(gmmest1)
```

```python
gmmest1.c_stat()
```

```python
gmmest = IVGMM((df3.LW), df3[["EXPR","TENURE" , "RNS",  "SMSA" , "YEAR_66" ,	"YEAR_67" ,	"YEAR_68" ,	"YEAR_69" ,	"YEAR_70"	, "YEAR_71" ,	"YEAR_73"]],df[["IQ","S"]],df3[["MED" ,"KWW","MRT","AGE"]]).fit(
    cov_type="unadjusted"
)
print(gmmest)
```
